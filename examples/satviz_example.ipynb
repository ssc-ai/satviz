# SatViz Jupyter Example

This notebook creates a SatSimJS widget, builds a small scenario in Python, and displays it in Jupyter.

```python
import json
import os

from satviz import SatSimJS

# For local development, serve the SatSimJS repo root with `npm run serve`.
SATSIM_BASE = os.getenv("SATSIM_BASE", "https://cdn.jsdelivr.net/npm/satsim@0.15.1/dist")
```

Build a scenario with one Haleakala ground observatory and one SGP4 satellite.

```python
scenario = {
    "simulationParameters": {
        "start_time": "2025-01-01T00:00:00Z",
        "end_time": "2025-01-01T06:00:00Z",
        "time_step": 60,
        "clock_step": "system_clock_multiplier",
        "clock_range": "loop_stop",
        "playback_state": "play",
    },
    "objects": [
        {
            "type": "GroundEOObservatory",
            "name": "Haleakala",
            "latitude": 20.706489,
            "longitude": -156.256921,
            "altitude": 3056,
            "height": 100,
            "width": 100,
            "y_fov": 5,
            "x_fov": 5,
            "field_of_regard": [],
        },
        {
            "type": "SGP4",
            "name": "ISS (ZARYA)",
            "tle1": "1 25544U 98067A   24270.51782528  .00008833  00000+0  16184-3 0  9991",
            "tle2": "2 25544  51.6425  60.8533 0005566 332.0513 164.5352 15.50162442429013",
            "orientation": "nadir",
            "color": "#00aaff",
        },
    ],
    "events": [
        {
            "time": 120,
            "type": "setGimbalAxes",
            "observer": "Haleakala",
            "axes": {"az": 45, "el": 30},
        },
        {
            "time": 600,
            "type": "trackObject",
            "observer": "Haleakala",
            "target": "ISS (ZARYA)",
        },
    ],
}
```

Display the widget. SatViz loads SatSimJS from the default CDN unless `satsim_base` is overridden.

```python
widget = SatSimJS(
    height="700px",
    satsim_base=SATSIM_BASE,
    scenario_data=json.dumps(scenario),
    viewer_options={
        "showWeatherLayer": False,
        "showNightLayer": False,
        "showLowResEarth": True,
    },
)
widget
```

Select the satellite by name after the widget has rendered.

```python
widget.selected_object = "ISS (ZARYA)"
```
